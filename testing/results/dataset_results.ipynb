```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams['font.size'] = 14
mpl.rcParams['lines.linewidth'] = 3
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.grid'] = True
mpl.rcParams['grid.alpha'] = 0.5
mpl.rcParams['figure.figsize'] = 10,6
```

```python
results_1 = np.load("data_type_results_1_5.npy")
results_2 = np.load("data_type_results_6_10.npy")
results_3 = np.load("data_type_results_11_16.npy")
results = np.concatenate([results_1, results_2, results_3])
results = results[results[:, 0] != 0]
```

```python
x = results[:, 0]
single_mean = results[:, 1]
single_diff = results[:, 2]
double_mean = results[:, 3]
double_diff = results[:, 4]
```

```python
plt.figure()
plt.plot(x[single_diff != 0], single_mean[single_diff != 0], label="Single Mean Data")
plt.plot(x[single_diff != 0], single_diff[single_diff != 0], label="Single Diff Data")
plt.plot(x, double_mean, label="Double Mean Data")
plt.plot(x[single_diff != 0], double_diff[single_diff != 0], label="Double Diff Data")

plt.ylabel("Accuracy")
plt.xlabel("Lookback Games")
plt.legend()
plt.show()
```
